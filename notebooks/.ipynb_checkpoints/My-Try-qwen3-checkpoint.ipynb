```python
import sys
sys.path.insert(0, "..")

import torch
import torch.nn as nn
import torch.nn.functional as F
from transformers import AutoTokenizer, AutoModelForCausalLM
from src.utils import get_device, print_model_summary, format_param_count

device = get_device()
print(f"Using device: {device}")
print(f"PyTorch version: {torch.__version__}")

# --- Qwen3-1.7B Config (hard-coded for clarity) ---
vocab_size        = 151_936
d_model           = 2048         # hidden_size
n_layer           = 28           # num_hidden_layers
num_heads         = 16           # num_attention_heads (Q heads)
num_kv_heads      = 8            # num_key_value_heads (K, V heads for GQA)
head_dim          = 128          # explicit head_dim (NOT d_model // num_heads!)
intermediate_size = 6144         # FFN intermediate (3x d_model for SwiGLU)
max_seq_len       = 40_960       # max_position_embeddings
rope_theta        = 1_000_000.0  # RoPE base frequency
rms_norm_eps      = 1e-6         # RMSNorm epsilon

# Derived
num_kv_groups = num_heads // num_kv_heads  # 16 // 8 = 2 (each KV head serves 2 Q heads)

# NOTE: head_dim=128, so d_model = num_heads * head_dim = 16 * 128 = 2048. Checks out.
# But KV projection size = num_kv_heads * head_dim = 8 * 128 = 1024. Much smaller!

text = "The Transformer architecture is important because"
```

```python
model = AutoModelForCausalLM.from_pretrained("Qwen/Qwen3-1.7B")
```

```python
import torch.nn.functional as F
import math


class RMSNorm(nn.Module):
    def __init__(self, dim, eps=1e-6):
        super().__init__()
        self.eps = eps
        self.weight = nn.Parameter(torch.ones(dim))  # γ, shape (dim,)

    def forward(self, x):
        # x: (..., dim)
        rms = x.pow(2).mean(dim=-1, keepdim=True).add(self.eps).sqrt()
        return x / rms * self.weight


def apply_rope(x, cos, sin):
    """Rotate x using precomputed cos/sin. x: (B, H, T, d_k)"""
    # Split last dim in half, rotate pairs of dimensions
    x1, x2 = x[..., :x.shape[-1]//2], x[..., x.shape[-1]//2:]
    rotated = torch.cat([-x2, x1], dim=-1)
    return x * cos + rotated * sin


def build_rope_cache(seq_len, head_dim, theta=1_000_000, device='cpu'):
    """Build the cos/sin tables for RoPE."""
    # positions: 0, 1, 2, ..., T-1
    positions = torch.arange(seq_len, device=device).float()
    # frequency for each pair of dimensions
    freqs = 1.0 / (theta ** (torch.arange(0, head_dim, 2, device=device).float() / head_dim))
    # outer product: (T, d_k/2)
    angles = torch.outer(positions, freqs)
    # repeat to match full head_dim: (T, d_k)
    angles = torch.cat([angles, angles], dim=-1)
    cos = angles.cos()[None, None, :, :]  # (1, 1, T, d_k)
    sin = angles.sin()[None, None, :, :]  # (1, 1, T, d_k)
    return cos, sin


class Qwen3Attention(nn.Module):
    def __init__(self, hidden_size, num_q_heads, num_kv_heads, head_dim):
        super().__init__()
        self.num_q_heads  = num_q_heads   # 16
        self.num_kv_heads = num_kv_heads  # 8
        self.head_dim     = head_dim      # 128
        self.groups       = num_q_heads // num_kv_heads  # 2 — each KV head serves 2 Q heads

        # Projections — all nn.Linear, no bias (matches Qwen3 config)
        self.q_proj = nn.Linear(hidden_size, num_q_heads  * head_dim, bias=False)  # (2048, 2048)
        self.k_proj = nn.Linear(hidden_size, num_kv_heads * head_dim, bias=False)  # (2048, 1024)
        self.v_proj = nn.Linear(hidden_size, num_kv_heads * head_dim, bias=False)  # (2048, 1024)
        self.o_proj = nn.Linear(num_q_heads * head_dim, hidden_size,  bias=False)  # (2048, 2048)

        # Qwen3-specific: per-head QK normalization
        self.q_norm = RMSNorm(head_dim)  # γ shape: (128,)
        self.k_norm = RMSNorm(head_dim)  # γ shape: (128,)

    def forward(self, x):
        B, T, _ = x.shape

        # ── 1. Project to Q, K, V ───────────────────────────────────────
        Q = self.q_proj(x)  # (B, T, 2048)
        K = self.k_proj(x)  # (B, T, 1024)
        V = self.v_proj(x)  # (B, T, 1024)

        # ── 2. Reshape into heads ───────────────────────────────────────
        Q = Q.view(B, T, self.num_q_heads,  self.head_dim).transpose(1, 2)  # (B, 16, T, 128)
        K = K.view(B, T, self.num_kv_heads, self.head_dim).transpose(1, 2)  # (B,  8, T, 128)
        V = V.view(B, T, self.num_kv_heads, self.head_dim).transpose(1, 2)  # (B,  8, T, 128)

        # ── 3. QK norm (Qwen3-specific) ─────────────────────────────────
        Q = self.q_norm(Q)  # (B, 16, T, 128)
        K = self.k_norm(K)  # (B,  8, T, 128)

        # ── 4. Apply RoPE ───────────────────────────────────────────────
        cos, sin = build_rope_cache(T, self.head_dim, device=x.device)
        Q = apply_rope(Q, cos, sin)
        K = apply_rope(K, cos, sin)

        # ── 5. GQA: expand K and V so every Q head has a matching KV ───
        # (B, 8, T, 128) → (B, 16, T, 128) by repeating each KV head 2x
        K = K.repeat_interleave(self.groups, dim=1)  # (B, 16, T, 128)
        V = V.repeat_interleave(self.groups, dim=1)  # (B, 16, T, 128)

        # ── 6. Scaled dot-product attention (causal) ────────────────────
        # F.scaled_dot_product_attention handles:
        #   scores = Q @ Kᵀ / √d_k
        #   causal mask (is_causal=True sets future tokens to -inf)
        #   softmax → α
        #   output = α @ V
        attn_out = F.scaled_dot_product_attention(Q, K, V, is_causal=True)
        # attn_out: (B, 16, T, 128)

        # ── 7. Merge heads and output projection ────────────────────────
        attn_out = attn_out.transpose(1, 2).contiguous().view(B, T, -1)  # (B, T, 2048)
        return self.o_proj(attn_out)  # (B, T, 2048)
```

```python
tokenizer = AutoTokenizer.from_pretrained("Qwen/Qwen3-1.7B")
input_tok = tokenizer(text, return_tensors="pt")['input_ids']
```

```python
embedding_layer = nn.Embedding(vocab_size, d_model)
W_E = embedding_layer(input_tok)
```

```python
rms_norm = nn.RMSNorm(d_model, eps=rms_norm_eps)
attn = Qwen3Attention(hidden_size=d_model, num_q_heads=num_heads, num_kv_heads=num_kv_heads, head_dim=head_dim)



res = W_E

W_E_norm = rms_norm(W_E)
W_attn = attn(W_E_norm) + res
```

```python
res = W_attn

gate_proj = nn.Linear(d_model, intermediate_size, bias=False)
up_proj = nn.Linear(d_model, intermediate_size, bias=False)  
down_proj = nn.Linear(intermediate_size, d_model, bias=False)

W_attn_norm = rms_norm(W_attn)
W_FFN = down_proj(F.silu(gate_proj(W_attn_norm)) * up_proj(W_attn_norm)) + res
W_FFN
```

```python
# Build all 28 transformer layers (persistent weights, created once)
layers = []
for _ in range(n_layer):
    layers.append({
        "rms_norm"  : nn.RMSNorm(d_model, eps=rms_norm_eps),
        "attn"      : Qwen3Attention(hidden_size=d_model, num_q_heads=num_heads, num_kv_heads=num_kv_heads, head_dim=head_dim),
        "ffn_norm"  : nn.RMSNorm(d_model, eps=rms_norm_eps),
        "gate_proj" : nn.Linear(d_model, intermediate_size, bias=False),
        "up_proj"   : nn.Linear(d_model, intermediate_size, bias=False),
        "down_proj" : nn.Linear(intermediate_size, d_model, bias=False),
    })

final_norm = nn.RMSNorm(d_model, eps=rms_norm_eps)
linear_out = nn.Linear(d_model, vocab_size, bias=False)
```

```python
# Load HF weights into our layers
sd = model.state_dict()

with torch.no_grad():
    embedding_layer.weight.copy_(sd["model.embed_tokens.weight"])

    for i in range(n_layer):
        p = f"model.layers.{i}"
        layers[i]["rms_norm"].weight.copy_(sd[f"{p}.input_layernorm.weight"])
        layers[i]["attn"].q_proj.weight.copy_(sd[f"{p}.self_attn.q_proj.weight"])
        layers[i]["attn"].k_proj.weight.copy_(sd[f"{p}.self_attn.k_proj.weight"])
        layers[i]["attn"].v_proj.weight.copy_(sd[f"{p}.self_attn.v_proj.weight"])
        layers[i]["attn"].o_proj.weight.copy_(sd[f"{p}.self_attn.o_proj.weight"])
        layers[i]["attn"].q_norm.weight.copy_(sd[f"{p}.self_attn.q_norm.weight"])
        layers[i]["attn"].k_norm.weight.copy_(sd[f"{p}.self_attn.k_norm.weight"])
        layers[i]["ffn_norm"].weight.copy_(sd[f"{p}.post_attention_layernorm.weight"])
        layers[i]["gate_proj"].weight.copy_(sd[f"{p}.mlp.gate_proj.weight"])
        layers[i]["up_proj"].weight.copy_(sd[f"{p}.mlp.up_proj.weight"])
        layers[i]["down_proj"].weight.copy_(sd[f"{p}.mlp.down_proj.weight"])

    final_norm.weight.copy_(sd["model.norm.weight"])
    linear_out.weight.copy_(sd["model.embed_tokens.weight"])  # weight tying

del sd
print("Weights loaded.")
```

```python
def Decoder(x):
    res = x
    for layer in layers:
        # Attention sub-block
        W_E_norm = layer["rms_norm"](res)
        W_attn   = layer["attn"](W_E_norm) + res
        res      = W_attn

        # FFN sub-block
        W_attn_norm = layer["ffn_norm"](res)
        W_FFN       = layer["down_proj"](F.silu(layer["gate_proj"](W_attn_norm)) * layer["up_proj"](W_attn_norm)) + res
        res         = W_FFN
    return res
```

```python
logits = linear_out(final_norm(Decoder(W_E)))
logits
```

```python
@torch.no_grad()
def generate(prompt, max_new_tokens=50):
    ids = tokenizer(prompt, return_tensors="pt")["input_ids"]
    for _ in range(max_new_tokens):
        W_E = embedding_layer(ids)
        logits = linear_out(final_norm(Decoder(W_E)))
        next_id = logits[0, -1, :].argmax().unsqueeze(0).unsqueeze(0)
        if next_id.item() == tokenizer.eos_token_id:
            break
        ids = torch.cat([ids, next_id], dim=1)
    return tokenizer.decode(ids[0], skip_special_tokens=True)

print(generate(text))
```
