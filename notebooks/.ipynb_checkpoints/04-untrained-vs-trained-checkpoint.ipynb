# Untrained vs Trained: Same Architecture, Worlds Apart

This notebook runs the **same input** through two versions of the same architecture:

| Model | What it is |
|-------|------------|
| **Untrained RQwen3** | Your from-scratch model with random weights (d_model=2048 to match real scale) |
| **Trained Qwen3-1.7B** | The real HuggingFace model with ~1.7B trained parameters |

Both models share the same core design: 28 transformer layers, 16 Q heads / 8 KV heads (GQA),
SwiGLU FFN, RoPE, RMSNorm. The only structural difference is that the real model also has
`q_norm` and `k_norm` (QK normalization) inside attention — your model doesn't have those yet.

**The point**: architecture is just a container. Training is what fills it with understanding.

### Memory note
Each model is ~3.4GB in float16. We load them **one at a time** and free memory between
phases, so peak usage stays around ~3.5GB.

```python
import sys
sys.path.insert(0, '..')

import gc
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import seaborn as sns
import numpy as np
import pandas as pd
from collections import OrderedDict
from transformers import AutoModelForCausalLM, AutoTokenizer

from src.rqwen3 import CoreConfig, RQwen3
from src.utils import get_device, format_param_count

sns.set_theme(style='whitegrid')
plt.rcParams.update({'figure.figsize': (14, 6), 'font.size': 11})
C = sns.color_palette('Set2', 8)  # C[0]=green, C[1]=orange, C[2]=blue, ...
UNTRAINED_COLOR = C[0]
TRAINED_COLOR = C[1]

device = get_device()
print(f'Device: {device}')
```

```python
# Same tokenizer and input for both models
MODEL_NAME = 'Qwen/Qwen3-1.7B'
SAMPLE_TEXT = 'Why is math so important?'

tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
input_ids = tokenizer(SAMPLE_TEXT, return_tensors='pt')['input_ids']
token_labels = tokenizer.convert_ids_to_tokens(input_ids[0].tolist())

print(f'Input: "{SAMPLE_TEXT}"')
print(f'Tokens ({len(token_labels)}): {token_labels}')
```

```python
def collect_weight_stats(model):
    """Collect weight statistics and subsampled data from any PyTorch model."""
    result = OrderedDict()
    for name, param in model.named_parameters():
        data = param.detach().cpu().float().flatten().numpy()
        step = max(1, len(data) // 10000)  # subsample for fast histograms
        result[name] = {
            'mean': float(data.mean()),
            'std': float(data.std()),
            'min': float(data.min()),
            'max': float(data.max()),
            'numel': len(data),
            'sample': data[::step],
        }
    return result


def categorize_weights(stats_dict):
    """Group subsampled weight data by component type."""
    cats = OrderedDict({
        'Embedding': [], 'Q Projection': [], 'K Projection': [],
        'V Projection': [], 'O Projection': [],
        'Gate Proj (FFN)': [], 'Up Proj (FFN)': [],
        'Down Proj (FFN)': [], 'Norm': [], 'LM Head': [],
    })
    for name, s in stats_dict.items():
        d = s['sample']
        if 'embed' in name:
            cats['Embedding'].append(d)
        elif 'q_proj' in name:
            cats['Q Projection'].append(d)
        elif 'k_proj' in name:
            cats['K Projection'].append(d)
        elif 'v_proj' in name:
            cats['V Projection'].append(d)
        elif 'o_proj' in name:
            cats['O Projection'].append(d)
        elif 'gate_proj' in name:
            cats['Gate Proj (FFN)'].append(d)
        elif 'up_proj' in name:
            cats['Up Proj (FFN)'].append(d)
        elif 'down_proj' in name:
            cats['Down Proj (FFN)'].append(d)
        elif 'lm_head' in name or 'head' in name:
            cats['LM Head'].append(d)
        else:
            cats['Norm'].append(d)
    return {k: np.concatenate(v) if v else np.array([]) for k, v in cats.items()}


def free_memory():
    """Force garbage collection and clear GPU cache."""
    gc.collect()
    if hasattr(torch.mps, 'empty_cache'):
        torch.mps.empty_cache()
    elif torch.cuda.is_available():
        torch.cuda.empty_cache()
```

---
## Phase 1: Analyze the Untrained Model

Create your RQwen3 at full Qwen3-1.7B scale (`d_model=2048` instead of your usual 512)
so the weight shapes match for a fair comparison. Collect everything we need, then free memory.

```python
# Create RQwen3 with Qwen3-1.7B dimensions
config_1_7b = CoreConfig({'d_model': 2048})
untrained = RQwen3(config_1_7b).half()  # float16 to match trained model

total_params = sum(p.numel() for p in untrained.parameters())
print(f'Untrained RQwen3: {total_params:,} params ({format_param_count(total_params)})')

# Weight stats
print('Collecting weight statistics...')
untrained_weight_stats = collect_weight_stats(untrained)
untrained_cats = categorize_weights(untrained_weight_stats)

# Forward pass on CPU (model is on CPU, input too)
print('Running forward pass...')
with torch.no_grad():
    u_logits = untrained(input_ids, store_attn=True)

# Save outputs
untrained_text = tokenizer.decode(u_logits.argmax(dim=-1)[0].cpu(), skip_special_tokens=True)

# Save attention maps (small: ~180KB total)
untrained_attn = []
for i in range(config_1_7b.n_layer):
    w = untrained.model_transformer.attns[i]._attn_weights
    if w is not None:
        untrained_attn.append(w.float())

print(f'Captured attention from {len(untrained_attn)} layers')
print(f'Output preview: "{untrained_text[:80]}..."')

# Free memory
del untrained, u_logits
free_memory()
print('Untrained model freed.')
```

---
## Phase 2: Analyze the Trained Model

Load the real Qwen3-1.7B from HuggingFace (~3.4GB download, cached after first time).
We use `attn_implementation='eager'` so it returns actual attention weight matrices.

```python
print('Loading trained Qwen3-1.7B...')
trained = AutoModelForCausalLM.from_pretrained(
    MODEL_NAME,
    torch_dtype=torch.float16,
    device_map='auto',
    attn_implementation='eager',  # required to extract attention weights
)
trained.eval()

total_params = sum(p.numel() for p in trained.parameters())
print(f'Trained Qwen3-1.7B: {total_params:,} params ({format_param_count(total_params)})')

# Weight stats
print('Collecting weight statistics...')
trained_weight_stats = collect_weight_stats(trained)
trained_cats = categorize_weights(trained_weight_stats)

# Forward pass with attention
trained_device = next(trained.parameters()).device
print('Running forward pass...')
with torch.no_grad():
    outputs = trained(
        input_ids.to(trained_device),
        output_attentions=True,
    )

# Greedy output (same decoding as untrained for fair comparison)
trained_text_greedy = tokenizer.decode(
    outputs.logits.argmax(dim=-1)[0].cpu(), skip_special_tokens=True
)

# Proper autoregressive generation
print('Generating text...')
gen_ids = trained.generate(
    input_ids.to(trained_device),
    max_new_tokens=128,
    do_sample=True,
    temperature=0.7,
    top_p=0.9,
    pad_token_id=tokenizer.eos_token_id,
)
trained_text_generated = tokenizer.decode(
    gen_ids[0][input_ids.shape[1]:].cpu(), skip_special_tokens=True
)

# Save attention maps
trained_attn = [a.float().cpu() for a in outputs.attentions]
print(f'Captured attention from {len(trained_attn)} layers')

# Free memory
del trained, outputs, gen_ids
free_memory()
print('Trained model freed.')
```

Same input, same architecture, same number of layers. One has random weights,
the other has been trained on trillions of tokens.

## Comparison 2: Weight Distributions — Random vs Learned

At initialization, all weights are drawn from simple distributions (Kaiming uniform for
Linear, normal for Embedding). Training reshapes these distributions as the model
learns which connections matter.

**What to look for**: trained weights often have heavier tails (some values grow large)
and may become asymmetric. The spread (std) also changes per component.

```python
def plot_weight_comparison(u_cats, t_cats):
    """Overlaid weight histograms: untrained vs trained."""
    components = ['Embedding', 'Q Projection', 'Gate Proj (FFN)',
                  'Down Proj (FFN)', 'O Projection', 'LM Head']

    fig, axes = plt.subplots(2, 3, figsize=(18, 10))
    axes = axes.flatten()

    for idx, comp in enumerate(components):
        ax = axes[idx]
        u_data = u_cats.get(comp, np.array([]))
        t_data = t_cats.get(comp, np.array([]))

        if len(u_data) > 0:
            ax.hist(u_data, bins=100, alpha=0.5, color=UNTRAINED_COLOR,
                    label='Untrained', density=True, edgecolor='none')
        if len(t_data) > 0:
            ax.hist(t_data, bins=100, alpha=0.5, color=TRAINED_COLOR,
                    label='Trained', density=True, edgecolor='none')

        ax.set_title(comp, fontsize=12, fontweight='bold')
        ax.set_xlabel('Weight Value')
        ax.set_ylabel('Density')
        ax.legend(fontsize=9)

    plt.suptitle('Weight Distributions: Untrained (green) vs Trained (orange)',
                 fontsize=15, fontweight='bold')
    plt.tight_layout()
    plt.show()


plot_weight_comparison(untrained_cats, trained_cats)
```

```python
# Bar chart: weight std by component
components = [c for c in untrained_cats if len(untrained_cats[c]) > 0]
u_stds = [untrained_cats[c].std() for c in components]
t_stds = [trained_cats[c].std() if len(trained_cats.get(c, [])) > 0 else 0 for c in components]

fig, ax = plt.subplots(figsize=(14, 6))
x = np.arange(len(components))
w = 0.35

ax.bar(x - w/2, u_stds, w, label='Untrained', color=UNTRAINED_COLOR, alpha=0.85)
ax.bar(x + w/2, t_stds, w, label='Trained', color=TRAINED_COLOR, alpha=0.85)

ax.set_xlabel('Component')
ax.set_ylabel('Weight Std (spread)')
ax.set_title('How Much Do Weights Spread Out? Untrained vs Trained',
             fontsize=14, fontweight='bold')
ax.set_xticks(x)
ax.set_xticklabels(components, rotation=25, ha='right')
ax.legend(fontsize=11)
plt.tight_layout()
plt.show()

# Table
print(f'{"Component":<20} {"Untrained Std":>14} {"Trained Std":>14} {"Ratio":>10}')
print('-' * 60)
for c, us, ts in zip(components, u_stds, t_stds):
    ratio = ts / us if us > 0 else float('inf')
    print(f'{c:<20} {us:>14.5f} {ts:>14.5f} {ratio:>9.2f}x')
```

## Comparison 3: Attention Patterns — Noise vs Understanding

Attention decides which tokens look at which other tokens. This is where the
difference between "random" and "trained" is most visible:

- **Untrained** (top row, blue): roughly uniform — every token attends equally to all previous tokens
- **Trained** (bottom row, red): structured patterns — heads specialize in different relationships

```python
def plot_attention_comparison(u_attn, t_attn, layer_idx, token_labels, heads=4):
    """Top row = untrained, bottom row = trained, same layer."""
    u = u_attn[layer_idx][0]  # (heads, seq, seq)
    t = t_attn[layer_idx][0]
    n = min(heads, u.shape[0], t.shape[0])
    short = [tok[:8] for tok in token_labels]

    fig, axes = plt.subplots(2, n, figsize=(4.5 * n, 9))
    if n == 1:
        axes = axes.reshape(2, 1)

    for i in range(n):
        # Untrained (top)
        sns.heatmap(
            u[i].numpy(), ax=axes[0][i], cmap='Blues', vmin=0,
            xticklabels=short, yticklabels=short if i == 0 else False,
            square=True, cbar=False,
        )
        axes[0][i].set_title(f'Head {i}', fontsize=11)
        if i == 0:
            axes[0][i].set_ylabel('UNTRAINED\n\nQuery', fontsize=11, fontweight='bold')
        axes[0][i].tick_params(axis='both', labelsize=7)
        axes[0][i].set_xticklabels([])

        # Trained (bottom)
        sns.heatmap(
            t[i].numpy(), ax=axes[1][i], cmap='Reds', vmin=0,
            xticklabels=short, yticklabels=short if i == 0 else False,
            square=True, cbar=False,
        )
        if i == 0:
            axes[1][i].set_ylabel('TRAINED\n\nQuery', fontsize=11, fontweight='bold')
        axes[1][i].set_xlabel('Key', fontsize=9)
        axes[1][i].tick_params(axis='both', labelsize=7)

    plt.suptitle(f'Attention Patterns -- Layer {layer_idx}',
                 fontsize=14, fontweight='bold')
    plt.tight_layout()
    plt.show()


# Early layer, middle layer, final layer
n_layers = min(len(untrained_attn), len(trained_attn))
for layer_idx in [0, n_layers // 2, n_layers - 1]:
    plot_attention_comparison(untrained_attn, trained_attn, layer_idx, token_labels)
```

```python
def plot_avg_attention_comparison(u_attn, t_attn, token_labels):
    """Side-by-side average attention (all heads, all layers)."""
    u_avg = torch.stack(u_attn).mean(dim=(0, 1, 2)).numpy()
    t_avg = torch.stack(t_attn).mean(dim=(0, 1, 2)).numpy()
    short = [tok[:10] for tok in token_labels]

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 7))

    sns.heatmap(u_avg, ax=ax1, cmap='Blues',
                xticklabels=short, yticklabels=short,
                square=True, annot=True, fmt='.3f', annot_kws={'size': 9})
    ax1.set_title('Untrained (Random Weights)', fontsize=13, fontweight='bold')
    ax1.set_xlabel('Key (attending to)')
    ax1.set_ylabel('Query (from)')

    sns.heatmap(t_avg, ax=ax2, cmap='Reds',
                xticklabels=short, yticklabels=short,
                square=True, annot=True, fmt='.3f', annot_kws={'size': 9})
    ax2.set_title('Trained (Qwen3-1.7B)', fontsize=13, fontweight='bold')
    ax2.set_xlabel('Key (attending to)')
    ax2.set_ylabel('Query (from)')

    plt.suptitle('Average Attention Across All Heads & Layers',
                 fontsize=15, fontweight='bold')
    plt.tight_layout()
    plt.show()

    print('Left: near-uniform (random). Right: structured (learned).')
    print('Notice how the trained model concentrates attention on specific tokens.')


plot_avg_attention_comparison(untrained_attn, trained_attn, token_labels)
```

## Comparison 4: Attention Entropy — How "Focused" Is Each Layer?

**Entropy** measures how spread out the attention is:
- **High entropy** = attention is spread evenly across all tokens (unfocused)
- **Low entropy** = attention is concentrated on a few tokens (focused)

Untrained models have high, uniform entropy. Trained models develop layers with
different levels of focus — some layers attend broadly (context gathering),
others attend narrowly (specific lookups).

```python
def attention_entropy(attn_list):
    """Compute mean entropy of attention distributions per layer."""
    entropies = []
    for attn in attn_list:
        # attn: (1, heads, seq, seq) -> probabilities along last dim
        p = attn[0]  # (heads, seq, seq)
        # Avoid log(0)
        log_p = torch.where(p > 0, torch.log2(p), torch.zeros_like(p))
        entropy = -(p * log_p).sum(dim=-1)  # (heads, seq)
        entropies.append(entropy.mean().item())  # average over heads and positions
    return entropies


u_entropy = attention_entropy(untrained_attn)
t_entropy = attention_entropy(trained_attn)

fig, ax = plt.subplots(figsize=(14, 5))
layers = range(len(u_entropy))

ax.plot(layers, u_entropy, 'o-', color=UNTRAINED_COLOR, label='Untrained',
        markersize=5, linewidth=2)
ax.plot(layers, t_entropy, 's-', color=TRAINED_COLOR, label='Trained',
        markersize=5, linewidth=2)

ax.set_xlabel('Layer', fontsize=12)
ax.set_ylabel('Mean Attention Entropy (bits)', fontsize=12)
ax.set_title('Attention Focus: Untrained vs Trained Across Layers',
             fontsize=14, fontweight='bold')
ax.legend(fontsize=12)
ax.fill_between(layers, u_entropy, t_entropy, alpha=0.1, color='gray')

plt.tight_layout()
plt.show()

print(f'Untrained avg entropy: {np.mean(u_entropy):.3f} bits (near-uniform)')
print(f'Trained avg entropy:   {np.mean(t_entropy):.3f} bits (more focused)')
print(f'Gap: {np.mean(u_entropy) - np.mean(t_entropy):.3f} bits')
print()
print('The gap = how much "focus" the model learned through training.')
```

## What I Learned

### Key Takeaways

**1. Training transforms noise into understanding**
- Same architecture, same number of parameters, completely different behavior
- The untrained model outputs random token sequences
- The trained model produces coherent, contextual text
- The architecture is a container; the weights are the knowledge

**2. Weight distributions shift dramatically**
- Initialization: symmetric, uniform spread
- After training: heavier tails, some weights grow large (important connections),
  many stay small (less important connections)
- Different component types learn different distributions

**3. Attention goes from noise to structured patterns**
- Untrained: near-uniform attention (every token attends equally to everything)
- Trained: heads specialize — some attend locally, some attend to semantic matches,
  some attend to special tokens
- Entropy quantifies this: trained attention is measurably more focused

**4. Different layers learn different roles**
- Early layers: broader attention (gathering context)
- Middle layers: mixing of local and semantic patterns
- Late layers: more focused attention (refining predictions)
- The trained entropy curve shows this layered specialization

**5. Minor architectural details matter**
- The real Qwen3-1.7B has `q_norm` and `k_norm` that your model doesn't
- These normalize Q and K per-head before attention, improving training stability
- Adding these to your RQwen3 is a good next step
