# Qwen3-1.7B from Scratch

Building every layer of the Qwen3-1.7B architecture by hand, bottom-up.

**Target model:** [`Qwen/Qwen3-1.7B`](https://huggingface.co/Qwen/Qwen3-1.7B) — 1.7B parameters, 28 layers, GQA, SwiGLU, RoPE.

| Component | Section |
|---|---|
| `Qwen3Config` | 1 |
| `Qwen3RMSNorm` | 2 |
| `Qwen3RotaryEmbedding` | 3 |
| `Qwen3MLP` | 4 |
| `Qwen3Attention` | 5 |
| `Qwen3DecoderLayer` | 6 |
| `Qwen3Model` | 7 |
| `Qwen3ForCausalLM` | 8 |

````python
# ─── 1. Qwen3Config ───────────────────────────────────────────────────────────
# Defaults match Qwen/Qwen3-1.7B exactly.
# head_dim is derived from hidden_size // num_attention_heads (not a free param).

@dataclass
class Qwen3Config:
    """Configuration for Qwen3 model architecture (defaults = Qwen3-1.7B)."""

    vocab_size: int = 151936
    hidden_size: int = 2048
    intermediate_size: int = 6144
    num_hidden_layers: int = 28
    num_attention_heads: int = 16
    num_key_value_heads: int = 8           # GQA: 8 KV heads shared across 16 Q heads
    hidden_act: str = "silu"
    max_position_embeddings: int = 40960
    rms_norm_eps: float = 1e-6
    rope_theta: float = 1_000_000.0        # 1M — much larger than the standard 10k
    attention_dropout: float = 0.0
    initializer_range: float = 0.02
    use_cache: bool = True
    tie_word_embeddings: bool = True        # 1.7B ties embed ↔ lm_head
    use_sliding_window: bool = False
    sliding_window: int = 4096
    max_window_layers: int = 28

    @property
    def head_dim(self) -> int:
        """Derived: hidden_size // num_attention_heads.  Not a free parameter."""
        return self.hidden_size // self.num_attention_heads


# Defaults are Qwen3-1.7B — no overrides needed
cfg = Qwen3Config()
print(cfg)
print(f"head_dim (derived): {cfg.head_dim}")
````

## 2. RMSNorm

Root Mean Square Layer Normalization — simpler and faster than LayerNorm.

$$\text{RMSNorm}(x) = \frac{x}{\sqrt{\text{mean}(x^2) + \epsilon}} \cdot \gamma$$

No mean-centering (no bias), just scale by learned $\gamma$ weights.

````python
# ─── 2. Qwen3RMSNorm ─────────────────────────────────────────────────────────

class Qwen3RMSNorm(nn.Module):
    """RMS Normalization (no mean-centering, just variance scaling)."""

    def __init__(self, hidden_size: int, eps: float = 1e-6):
        super().__init__()
        self.weight = nn.Parameter(torch.ones(hidden_size))  # learnable γ
        self.eps = eps

    def forward(self, hidden_states: torch.Tensor) -> torch.Tensor:
        # TODO: implement
        # 1. Cast to float32 for numerical stability
        # 2. Compute x * rsqrt(mean(x^2) + eps)
        # 3. Scale by self.weight, cast back to input dtype
        raise NotImplementedError


# ── smoke test ──
norm = Qwen3RMSNorm(cfg.hidden_size)
x = torch.randn(2, 8, cfg.hidden_size)
# norm(x).shape  # uncomment after implementing → (2, 8, 2048)
````

## 4. MLP (Feed-Forward Network)

Qwen3 uses a **SwiGLU** variant — three projections instead of two:

$$\text{MLP}(x) = W_\text{down} \cdot \big[\text{SiLU}(W_\text{gate} \cdot x) \odot (W_\text{up} \cdot x)\big]$$

Expansion ratio: `intermediate_size / hidden_size` = 6144/2048 = **3x**

````python
# ─── 4. Qwen3MLP ─────────────────────────────────────────────────────────────

class Qwen3MLP(nn.Module):
    """Gated feed-forward (SwiGLU): three linear projections, no bias."""

    def __init__(self, config: Qwen3Config):
        super().__init__()
        self.hidden_size = config.hidden_size
        self.intermediate_size = config.intermediate_size

        self.gate_proj = nn.Linear(config.hidden_size, config.intermediate_size, bias=False)
        self.up_proj   = nn.Linear(config.hidden_size, config.intermediate_size, bias=False)
        self.down_proj = nn.Linear(config.intermediate_size, config.hidden_size, bias=False)

    def forward(self, x: torch.Tensor) -> torch.Tensor:
        """x: (batch, seq_len, hidden_size) → same shape."""
        # TODO: implement
        # return self.down_proj(F.silu(self.gate_proj(x)) * self.up_proj(x))
        raise NotImplementedError


# ── smoke test ──
mlp = Qwen3MLP(cfg)
print(f"MLP params: gate={cfg.hidden_size}→{cfg.intermediate_size}, "
      f"up={cfg.hidden_size}→{cfg.intermediate_size}, "
      f"down={cfg.intermediate_size}→{cfg.hidden_size}")
# mlp(x).shape  # → (2, 8, 2048)
````

## 5. Attention

Key Qwen3 differences from vanilla multi-head attention:
1. **Grouped Query Attention (GQA)** — K/V can have fewer heads than Q

4. **RoPE** applied after projection + norm

## 6. Decoder Layer

One full transformer block — **pre-norm** architecture:

```
         x
         │
    ┌────▼────┐
    │ RMSNorm │
    └────┬────┘
         │
    ┌────▼─────┐
    │ Attention │
    └────┬─────┘
         │
    x + attn_out   ← residual
         │
    ┌────▼────┐
    │ RMSNorm │
    └────┬────┘
         │
    ┌────▼───┐
    │  MLP   │
    └────┬───┘
         │
    x + mlp_out    ← residual
         │
         ▼
```

````python
# ─── 6. Qwen3DecoderLayer ─────────────────────────────────────────────────────

class Qwen3DecoderLayer(nn.Module):
    """Single transformer block: pre-norm attention + pre-norm MLP with residuals."""

    def __init__(self, config: Qwen3Config, layer_idx: int = 0):
        super().__init__()
        self.self_attn = Qwen3Attention(config, layer_idx)
        self.mlp = Qwen3MLP(config)
        self.input_layernorm = Qwen3RMSNorm(config.hidden_size, eps=config.rms_norm_eps)
        self.post_attention_layernorm = Qwen3RMSNorm(config.hidden_size, eps=config.rms_norm_eps)

    def forward(
        self,
        hidden_states: torch.Tensor,
        position_embeddings: Tuple[torch.Tensor, torch.Tensor],
        attention_mask: Optional[torch.Tensor] = None,
    ) -> torch.Tensor:
        """
        hidden_states: (batch, seq_len, hidden_size)
        Returns:       (batch, seq_len, hidden_size)
        """
        # TODO: implement
        # 1. residual = hidden_states
        # 2. hidden_states = self.input_layernorm(hidden_states)
        # 3. hidden_states = self.self_attn(hidden_states, position_embeddings, attention_mask)
        # 4. hidden_states = residual + hidden_states          ← first residual
        # 5. residual = hidden_states
        # 6. hidden_states = self.post_attention_layernorm(hidden_states)
        # 7. hidden_states = self.mlp(hidden_states)
        # 8. hidden_states = residual + hidden_states          ← second residual
        # 9. return hidden_states
        raise NotImplementedError


# ── smoke test ──
layer = Qwen3DecoderLayer(cfg, layer_idx=0)
total = sum(p.numel() for p in layer.parameters())
print(f"Decoder layer params: {total:,}")
````

## 7. Qwen3Model — the "bare" transformer

Stacks everything together: embedding → N decoder layers → final norm.

RoPE is computed **once** here and shared across all layers (not per-layer).

````python
# ─── 7. Qwen3Model (bare transformer) ─────────────────────────────────────────

class Qwen3Model(nn.Module):
    """Bare Qwen3 transformer: embed → layers → norm.  No LM head."""

    def __init__(self, config: Qwen3Config):
        super().__init__()
        self.config = config

        self.embed_tokens = nn.Embedding(config.vocab_size, config.hidden_size)
        self.layers = nn.ModuleList([
            Qwen3DecoderLayer(config, layer_idx=i)
            for i in range(config.num_hidden_layers)
        ])
        self.norm = Qwen3RMSNorm(config.hidden_size, eps=config.rms_norm_eps)
        self.rotary_emb = Qwen3RotaryEmbedding(config)

    def forward(
        self,
        input_ids: torch.LongTensor,
        attention_mask: Optional[torch.Tensor] = None,
        position_ids: Optional[torch.LongTensor] = None,
    ) -> torch.Tensor:
        """
        input_ids:    (batch, seq_len)
        Returns:      (batch, seq_len, hidden_size)  — last hidden state
        """
        # TODO: implement
        # 1. hidden_states = self.embed_tokens(input_ids)
        #
        # 2. Build position_ids if not provided:
        #    position_ids = torch.arange(seq_len, device=device).unsqueeze(0)
        #
        # 3. Compute RoPE once (shared across all layers):
        #    cos, sin = self.rotary_emb(hidden_states, position_ids)
        #    position_embeddings = (cos, sin)
        #
        # 4. Build causal attention mask if needed
        #
        # 5. Pass through each decoder layer:
        #    for layer in self.layers:
        #        hidden_states = layer(hidden_states, position_embeddings, attention_mask)
        #
        # 6. Final norm:
        #    hidden_states = self.norm(hidden_states)
        #
        # 7. return hidden_states
        raise NotImplementedError


# ── smoke test ──
model = Qwen3Model(cfg)
total = sum(p.numel() for p in model.parameters())
print(f"Qwen3Model (bare) total params: {total:,}")
print(f"  Embedding: {cfg.vocab_size} × {cfg.hidden_size} = {cfg.vocab_size * cfg.hidden_size:,}")
print(f"  Layers: {cfg.num_hidden_layers}")
print(f"  Final norm: {cfg.hidden_size}")
````

## 8. Qwen3ForCausalLM — the full language model

Wraps `Qwen3Model` and adds:
- **LM head**: `nn.Linear(hidden_size → vocab_size)` to produce logits
- **Loss**: cross-entropy on shifted labels (predict next token)

We define a small `CausalLMOutput` dataclass so the return type is typed and IDE-friendly (instead of a raw `dict`).

````python
# ─── 8. Qwen3ForCausalLM ──────────────────────────────────────────────────────

@dataclass
class CausalLMOutput:
    """Typed return value for Qwen3ForCausalLM.forward()."""
    logits: torch.Tensor                    # (batch, seq_len, vocab_size)
    loss: Optional[torch.Tensor] = None     # scalar, only when labels provided


class Qwen3ForCausalLM(nn.Module):
    """Qwen3 for causal language modeling: backbone + LM head + loss."""

    def __init__(self, config: Qwen3Config):
        super().__init__()
        self.config = config
        self.model = Qwen3Model(config)
        self.lm_head = nn.Linear(config.hidden_size, config.vocab_size, bias=False)

        # Qwen3-1.7B ties embedding ↔ lm_head weights (saves ~311M params)
        if config.tie_word_embeddings:
            self.lm_head.weight = self.model.embed_tokens.weight

    def forward(
        self,
        input_ids: torch.LongTensor,
        attention_mask: Optional[torch.Tensor] = None,
        position_ids: Optional[torch.LongTensor] = None,
        labels: Optional[torch.LongTensor] = None,
    ) -> CausalLMOutput:
        """
        input_ids: (batch, seq_len)
        labels:    (batch, seq_len)  — shifted internally for next-token prediction
        Returns:   CausalLMOutput with .logits and optionally .loss
        """
        # TODO: implement
        # 1. hidden_states = self.model(input_ids, attention_mask, position_ids)
        #
        # 2. logits = self.lm_head(hidden_states)   → (batch, seq_len, vocab_size)
        #
        # 3. If labels provided, compute cross-entropy loss:
        #    - Shift: logits[..., :-1, :] vs labels[..., 1:]
        #    - loss = F.cross_entropy(shift_logits.view(-1, vocab_size), shift_labels.view(-1))
        #
        # 4. return CausalLMOutput(logits=logits, loss=loss)
        raise NotImplementedError


# ── smoke test ──
causal_lm = Qwen3ForCausalLM(cfg)
total = sum(p.numel() for p in causal_lm.parameters())
print(f"Qwen3ForCausalLM total params: {total:,}")
print(f"  LM head: {cfg.hidden_size} → {cfg.vocab_size} (tied={cfg.tie_word_embeddings})")
````
